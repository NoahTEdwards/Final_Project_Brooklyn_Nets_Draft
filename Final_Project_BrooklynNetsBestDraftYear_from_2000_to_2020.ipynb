### The Data

Here I am loading in the csv file that contains all of the data for the brooklyn nets draft years.

```python
bn_draft = pd.read_csv(r'C:\Users\ntedw\EMAT 22110\Brooklyn_Nets_Draft.csv')
bn_draft.keys()
```

I am loading in all of the data.

```python
bn_draft.head(50)
```

Here I am creating Years and PPG to combine them together and make one column.

```python
Years = bn_draft['Year']
Years.head()
```

```python
PPG = bn_draft['PTS Per Game']
PPG.head()
```

Here I combine Years and PPG to create Years_PPG.

```python
Years_PPG = bn_draft[['Year','PTS Per Game']]
```

 ### Visualization PTS Per Game

In my visualization you will be able to see two charts. One chart being a chart with the years in descending order and the other chart being a chart with the Points Per Game in descending order.

I will also be showing one graph with the years in descending order. Each year will show the average Points Per Game. 


### PTS Per Game Chart 1

This chart is showing the Points Per Game in descending order.

```python
Years_PPGs = Years_PPG.sort_values('PTS Per Game', ascending = False)[:75]
Years_PPGs
```

### PTS Per Game Chart 2

This chart is showing the Years in descending order.

```python
Years_PPG = bn_draft[['Year','PTS Per Game']]
Years_PPG.head(40)
```

### PTS Per Game Graph 

This graph shows the years in descending order and also shows what year had the highest Points Per Game.

```python
fig, ax0 = plt.subplots(figsize=(15,10))
ax0.bar(Years, Years_PPG['PTS Per Game'])

plt.xticks(Years,rotation=90, fontsize = 'small')
plt.xlabel('Years')
plt.ylabel('Points Per Game')
plt.title("Brooklyn Nets Draft: Points Per Game Every Year ")
plt.show()
```

### Visualization Three Point Percentage
